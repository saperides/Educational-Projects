I'm going to remove encounter_id as this is an arbitrary assignment as well as weight, because we only have data on 3% of the patients. This is a shame because I believe this could hold valuable insight, but I don't see a way around it.
I'll use the mean to fill missing values in diag_1, diag_2, and diag_3. 
Finally, I'll create dummies for all columns aside from time_in_hospital, num_procedures, num_medications, diag_1, diag_2, diag_3, and num_diagnoses. <br>
A note: I don't know what diag_1, diag_2, and diag_3 represent. My initial thought was that they are diagnoses 1, 2, and 3, so the values ought to be categorical. However, there are a few float values indicating that the information is in fact numerical/continuous so I will treat it that way.

```python
df.drop(columns=['encounter_id', 'weight'], inplace=True)
for df_column in ['diag_1', 'diag_2', 'diag_3']:
        df[df_column].fillna(df[df_column].mean, inplace=True)
             
df.fillna('Uknown', inplace=True)
df['race'].value_counts()
df.info()
```

```python
# pd.options.display.max_rows = None
# df.diag_2.value_counts()[47]
df[df['diag_2'].apply(lambda x: isinstance(x, str))]
```

```python
# for column in ['diag_1', 'diag_2', 'diag_3']:
# #     print('\n\n', df[column].value_counts())
#     print(df[column].dtype)
#     df[column] = df[column].apply(lambda x: float(x))
#     print(df[column].dtype)
    
df['diag_2'] = pd.to_numeric(df['diag_2'])
```

```python
# Getting dummies for categorical variables
for column in ['race', 'gender', 'age', 'payer_code', 'medical_specialty', 'max_glu_serum', 
               'A1Cresult', 'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide',
               'glipizide','glyburide','tolbutamide','pioglitazone','rosiglitazone', 'acarbose','miglitol','troglitazone',
               'tolazamide','examide','citoglipton','insulin','glyburide-metformin','glipizide-metformin',
               'glimepiride-pioglitazone','metformin-rosiglitazone','metformin-pioglitazone','change','diabetesMed','readmitted']:
#     print('column is {}'.format(column))
    df = pd.concat([df, pd.get_dummies(df[column])], axis=1)
```
